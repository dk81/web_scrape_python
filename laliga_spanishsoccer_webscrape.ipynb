## All In One Function

```python
def scrape_laliga_table():

    url = "https://www.laliga.com/en-GB/laliga-santander/standing"
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')

    results = soup.find('div', 
            {'class': 'styled__StandingTableBody-e89col-5 cDiDQb'}).find_all('div', 
            {'class': 'styled__ContainerAccordion-e89col-11 HquGF'})
    
    # Scraping La Liga Table with list comprehension
    teams = [result.find('div', {'class': 'styled__ShieldContainer-lo8ov8-0 bkblFd shield-desktop'}).find('p', 
                        {'class': 'styled__TextRegularStyled-sc-1raci4c-0 glrfl'}).get_text() 
                        for result in results]

    points = [result.find_all('p', {'class': 'styled__TextRegularStyled-sc-1raci4c-0 cIcTog'})[0].get_text()
              for result in results]

    played = [result.find_all('p', {'class': 'styled__TextRegularStyled-sc-1raci4c-0 cIcTog'})[1].get_text()
             for result in results]

    wins = [result.find_all('p', {'class': 'styled__TextRegularStyled-sc-1raci4c-0 cIcTog'})[2].get_text()
            for result in results]

    draws = [result.find_all('p', {'class': 'styled__TextRegularStyled-sc-1raci4c-0 cIcTog'})[3].get_text()
             for result in results]

    losses = [result.find_all('p', {'class': 'styled__TextRegularStyled-sc-1raci4c-0 cIcTog'})[4].get_text()
             for result in results]

    goals_for = [result.find_all('p', {'class': 'styled__TextRegularStyled-sc-1raci4c-0 cIcTog'})[5].get_text()
             for result in results]

    goals_against = [result.find_all('p', {'class': 'styled__TextRegularStyled-sc-1raci4c-0 cIcTog'})[6].get_text()
                     for result in results]

    goals_diff = [result.find_all('p', {'class': 'styled__TextRegularStyled-sc-1raci4c-0 cIcTog'})[7].get_text()
                 for result in results]
    
    df = pd.DataFrame({'Rank': range(1, len(teams) + 1), 'Team': teams, 'Points': points,
                       'Played': played, 'Wins': wins, 'Draws': draws, 'Losses': losses, 
                       'Goals For': goals_for, 'Goals Against': goals_against, 'Goal Difference': goals_diff})
    
    return df
```

```python
# Function Call:

laliga_table = scrape_laliga_table()

laliga_table
```
